```python
"""This program plays a game of Rock, Paper, Scissors between two Players,
and reports both Player's scores each round."""
import pandas as pd
import random
moves = ['rock', 'paper', 'scissors']

"""The Player class is the parent class for all of the Players
in this game"""


class Player:
    def move(self):
        return 'rock'

    def learn(self, my_move, their_move):
        pass
    
class Random_Player(Player):
    def move(self):
        return (random.choice(moves))

    def learn(self, my_move, their_move):
        pass
    

class Reflect_Player(Player):
    def __init__(self):
        Player.__init__(self)
        self.their_move = None

    def learn(self, my_move, their_move):
        self.their_move = their_move

    def move(self):
        if self.their_move == None:
            return (random.choice(moves))
        else:
            return self.their_move


def beats(one, two):
    return ((one == 'rock' and two == 'scissors') or
            (one == 'scissors' and two == 'paper') or
            (one == 'paper' and two == 'rock'))


class Game:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        self.count_wins = 0
        self.count_losses = 0
        self.count_ties = 0

    def play_round(self):
        move1 = self.p1.move()
        move2 = self.p2.move()
        print(f"Player 1: {move1}  Player 2: {move2}")
        self.p1.learn(move1, move2)
        self.p2.learn(move2, move1)

    def play_game(self):
        move1 = self.p1.move()
        move2 = self.p2.move()
        print(f"Player 1 Move: {move1}  Player 2 Move: {move2}")

        if beats(move1, move2):
            self.count_wins += 1
            print(f"wins:{self.count_wins}")
        elif move1 == move2:
            self.count_ties += 1
            print(f"ties:{self.count_ties}")
        elif beats(move2, move1):
            self.count_losses += 1
            
        self.score1 = self.count_wins
        self.score2 = self.count_wins
        
        
            
            
        print("Game start!")
        for round in range(10):
            print(f"Round {round}:")
            self.play_round()
            
        data = {'Player':['Player1','Player2'],'Wins':[self.score1,self.score2]}
        df = pd.DataFrame(data)
        print(df)
        print("Game over!")


if __name__ == '__main__':
    game = Game(Player(), Random_Player())
    game.play_game()
```
