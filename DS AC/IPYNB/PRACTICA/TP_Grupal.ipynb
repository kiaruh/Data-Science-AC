A value linked to when the review was entered in the database. Higher = more recent.
Review Date

Date of publication of the review.
Cocoa Percent

Cocoa percentage (darkness) of the chocolate bar being reviewed.
Company Location

Manufacturer base country.
Rating

Expert rating for the bar.
Bean Type

The variety (breed) of bean used, if provided.
Broad Bean Origin

The broad geo-region of origin for the bean.

Para simplificar el trabajo vamos a cambiar el nombre de las columnas. 

```python
df.columns = ['Company', 'Specific Bean Origin or Bar Name', 'REF','Review Date','Cocoa Percent','Location', 'Rating','Bean Type', 'Bean Origin']
```

```python
df.head(2)
```

Vemos que en la columna Cocoa Percent esta el simbolo de porcentaje, que no nos ayuda para graficar el dataframe. 

```python
df['Company']=df['Company'].astype('category')
```

```python
df['Cocoa Percent'] = df['Cocoa Percent'].str.replace('%', ' ')
df['Cocoa Percent'] = df['Cocoa Percent'].astype(float)
```

```python
df.describe(include='all').T
```

```python
import seaborn as sns
sns.countplot(x='Rating', data=df)
plt.xlabel('Rating')
plt.ylabel('Count of Users')
plt.title('Number of Users that Rated Chocolate Bars')
print('Fig 1: Count of Chocolate Bar Ratings')
```
