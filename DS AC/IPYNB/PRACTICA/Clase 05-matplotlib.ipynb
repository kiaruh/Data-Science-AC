#### Importamos la librería y seteamos que matplotlib inserte el gráfico en el jupyter notebook

```python
import numpy as np
import matplotlib.pyplot as plt
```

El gráfico más simple es cuando queremos visualizar una función $y = f(x)$.
Veamos como crearlo.

#### Simple line plot

Empezamos creando una figura y asignándole ejes

```python

fig = plt.figure()
ax = plt.axes()
plt.show()
```

Llamamos a la función universal para que muestre el gráfico dentro del jupyter notebook

```python
%matplotlib inline
```

En Matplotlib, la *figure* (``plt.Figure``) se puede pensar como un contenedor de  todos los objetos que representan los ejes, el texto, las etiquetas, el gráfico. Los *axes* (``plt.Axes``)  es lo que vemos, una caja con ticks y labels, que después va a contener los elementos del gráfico que hacen nuestra visualización

Con la función ``ax.plot`` podemos graficos datos. 

Empecemos con un ejemplo bien simple.
Grafiquemos una función sinusoidal.

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

Podemos usar la interfase de pylab donde la figura y los ejes son creadas para nosotros
(pueden ver [Two Interfaces for the Price of One](04.00-Introduction-To-Matplotlib.ipynb#Two-Interfaces-for-the-Price-of-One)):

Si queremos hacer un gráfico con muchas líneas

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

Para que el gráfico pueda ser interpretado es necesario decir qué estamos graficando!
Para esto agreguemos una etiqueta en los ejes x e y y agreguemos una leyenda a cada línea.
También podemos darle un título.

```python
plt.plot(x, np.sin(x),label='sin(x)')
plt.plot(x, np.cos(x),label='cos(x)');
plt.xlabel('x')
plt.ylabel('y')
plt.title("Funciones seno y coseno")
plt.legend()
plt.show()
```

Si queremos ajustar el color o grosor de la línea especificamos en el argumento ``color`` y ``linestyle``

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

#### Scatter plot

```python
testscores_class1=[35, 50, 65, 82, 90, 99]

timestudying= [1,3, 4,6, 7, 8]

plt.scatter(timestudying, testscores_class1)
plt.title('Student Test Grades')
plt.xlabel('Time Spent Studying')
plt.ylabel('Test Scores')
plt.show()
```

##### Si queremos superponer dos conjuntos de datos tenemos que asignar un código de color

```python
testscores_class1=[35, 50, 65, 82, 90, 99]

testscores_class2=[85, 90, 92, 95, 97, 100]

timestudying= [1,3, 4,6, 7, 8]

fig = plt.figure()
ax = plt.axes()
plt.scatter(timestudying, testscores_class1, label='Class 1')
plt.scatter(timestudying, testscores_class2, label='Class 2')

plt.title('Student Test Grades')
plt.xlabel('Time Spent Studying')
plt.ylabel('Test Scores')
plt.legend() 
plt.show()
```

En principio ambas funciones son equivalentes pero ``plt.plot`` puede ser mucho más eficiente que  ``plt.scatter`` cuando tenemos muchos datos

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

#### Gráficos con subfiguras

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

Esta interfase tienen la dificultad de que una vez que accedemos al segundo panel  no podemos volver a modificar nada del primero.

Para resolver esto podemos usar una interfase basada en objetos que nos permite tener más control sobre la figura:

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

#### Visualización de los errores

Las variables medidas tienen errores y es importante, además del valor medido, asignarles un error.

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

 ``fmt`` es un código de formato que controla la apariencia de líneas y puntos (para más información leer [Simple Line Plots](04.01-Simple-Line-Plots.ipynb) y [Simple Scatter Plots](04.02-Simple-Scatter-Plots.ipynb)).
