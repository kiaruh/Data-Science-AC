```python
from operator import attrgetter

class cuentas:
    numerojugador=0
    nombre=""
    puesto=0
    goles=0
    
    def __init__(self,numerojugador,nombre,puesto,goles):
        self.numerojugador=numerojugador
        self.nombre=nombre
        self.puesto=puesto
        self.goles=goles
    
def mostrar(jugador):
    for i in range (len(jugador)):
        print("\nNumero de jugar: {}\nNombre: {}\nPuesto: {}\nGoles: {}\n".format(
            jugador[i].numerojugador,jugador[i].nombre,jugador[i].puesto,jugador[i].goles))
        
        
def goles(jugador):
    positivo=0
    cero=0
    for i in range (len(jugador)):
        if jugador[i].saldo > 0:
            positivo+=1
        elif jugador[i].saldo == 0:
            cero+=1
    print("\nHay {} que metieron algun gol y {} que no metieron ningun go\n" .format(positivo,cero))
        
        
        
###def topcinco(jugador): # vemos
#    goles=jugador[0].goles
#    nombre=jugador[0].nombre
#    puesto=jugador[0].puesto
#    for i in range (len(jugador)):
#        if goles > jugador[i].goles:
#            nombre=jugador[i].nombre
#            puesto=jugador[i].puesto
#    print()
    

def cargar(jugador):
    numerojugador=int(input("Ingrese numero jugador: "))
    while numerojugador < 0:
        print("Dato incorrecto")
        numerojugador=int(input("ingrese numero jugador: "))
    while numerojugador != 0:
        nombre=input("Ingrese el nombre: ")
        puesto=int(input("Ingrese 1. Arquero, 2. Defensa 3. Medio campista 4. Delantero : "))
        while not (puesto == 1 or puesto == 2 or puesto == 3 or puesto == 4 ):
            print("Dato incorrecto")
            puesto=int(input("Ingrese 1. Arquero, 2. Defensa 3. Medio campista 4. Delantero : "))
        goles=int(input("goles: "))
        jugador.append(cuentas(numerojugador,nombre,puesto,goles))
        numerojugador=int(input("ingrese numero jugador: "))
        while numerojugador < 0:
            print("Dato incorrecto")
            numerojugador=int(input("ingrese numero jugador: "))
        

def main():
    jugador=[]
    cargar(jugador)
    if len(jugador) == 0:
        print("no se ingresaron datos")
    else:
        #saldos(jugador)
        #topcinco(jugador)
        jugador.sort(key=attrgetter("goles"))
        mostrar(jugador)
        
```

```python
main()
```

```python
arreglo = []
arregloSort = []
pos = []
neg = []
def cargar():
    while len(arreglo)!=10:
        a = int(input("Cargar numero: "))
        arreglo.append(a)
def mayor():
    arreglo.sort(reverse=True)
    for i in range(0,4):
        arregloSort.append(arreglo[i])
    arregloSort.sort(reverse=True)
    
def generar():
    for i in arreglo:
        if i > 0 :
            pos.append(i)
        elif i < 0 :
            neg.append(i)
            
cargar()
print("arreglo cargado")
print (arreglo)
generar()
mayor()
print(arregloSort)
print(pos)
print(neg)

```

```python
NomCiudad = []
DiaVuelo = []
CapAvion = []
CantVentPas = []
CapacidadMaxima = []

#ciudad el dia con mayor cantidad de pasajes vendidos
#el porcentaje de pasajes vendidos de cada ciudad
#gerenar un arreglo con el nombre de la ciudad cuyo vuelvo no tenga mas asientos libres

def cargar():
    a = input("queres si ingresar \nSI para ingresar \nFIN para finalizar\n")
    while a != "FIN":
        if a == "SI":
            NC = input("Ingrese Nombre ciudad: ")
            DV = input("Ingrese dia de vuelvo dd/mm/aa: ")
            CA = int(input("Ingrese la capacidad del avion: "))
            CV = int(input("Ingrese la cantidad de ventas: "))
            NomCiudad.append(NC)
            DiaVuelo.append(DV)
            CapAvion.append(CA)
            CantVentPas.append(CV)
            a = input("queres seguir ingresando? \nSI para ingresar \nFIN para finalizar\n")
        else:
            print("error volve a ingresar")
            a = input("queres si ingresar \nSI para ingresar \nFIN para finalizar\n")
        
def mostrar_max():
    aux3 = 0
    aux4 = 0
    for i in range(len(NomCiudad)):
        if CantVentPas[i] > aux3:
            aux3 = CantVentPas[i]
            aux4 = i
    print("DESTINO CON MAS VENTAS DE PASAJE:")
    print("Nombre de la ciudad: {}" .format(NomCiudad[aux4]))
    print("Cantidad de ventas: {}" .format(CantVentPas[aux4]))
    
def mostrar_maxima_capacidad():
    for i in range(len(NomCiudad)):
        if CapAvion[i] == CantVentPas[i]:
            CapacidadMaxima.append(NomCiudad[i])
    print("Destinos cuyo vuelo llegaron a su capacidad maxima son: ")
    print(CapacidadMaxima)
            
def mostrar():
    aux1 = 0
    for i in range(len(NomCiudad)):
        aux1 += CantVentPas[i]
    for i in range(len(NomCiudad)):
        print("Nombre de la ciudad: {}" .format(NomCiudad[i]))
        print("Fecha: {}" .format(DiaVuelo[i]))
        print("Capacidad del Avion: {}" .format(CapAvion[i]))
        print("Cantidad de ventas: {}" .format(CantVentPas[i]))
        aux5 = CantVentPas[i]/aux1
        print("El porcentaje sobre: {:.1%}" .format(aux5))
        
              
cargar()
mostrar()
mostrar_maxima_capacidad()
mostrar_max()
```
