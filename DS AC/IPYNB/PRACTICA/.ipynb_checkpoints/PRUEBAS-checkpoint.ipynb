```python
#cargar un arreglo de 100 elementos y mostrar
#cuantos numeros pares mayor a promedio
arreglo = []
arreglo2 = []


def ingresar(arreglo):
    print("Iniciando ingreso de numero:\n")
    for i in range(0,10):
        ingresado = int(input("Ingrese numero: "))
        arreglo.append(ingresado)
    print("\nFinalizado el ingreso")

def mostrar_pares_mayor_prom(arreglo):
    total_numero = 0
    for num in arreglo:
        total_numero += num
    prom = total_numero/10
    for num in arreglo:
        if num%2 == 0 and num > prom:
            arreglo2.append(num)
    print("El promedio es : {}".format(prom))
    print("\nNumeros mayores al prom y pares: ")
    print(arreglo2)
    

def main():
    ingresar(arreglo)
    print(arreglo)
    mostrar_pares_mayor_prom(arreglo)

main()
```

```python
#sistema para articulos
#codigo producto de 1 a 50
#cantidad vendida
#tipo de pago E efectivo T tarjeta
#finalizar la carga, mostrar ventas totales del dia
#producto q mas se vendio
#promedio de venta
#porcentaje de ventas hecho en efectivo

def main():
    ventas = []
    cargar(ventas)
    if len(ventas)==0:
        print("no se ingreso datos")
    print("Mostrar ventas totales del dia")
    mostrar(ventas)
    mostrar_mas_vendido(ventas)
    mostrar_promedio_venta(ventas)
    mostrar_venta_en_efectivo(ventas)

class venta:
    tipo_producto = ""
    numero_de_venta = 0
    precio = 0
    tipo_pago = ""
    def __init__(self,tipo_producto,numero_de_venta,precio,tipo_pago):
        self.tipo_producto = tipo_producto
        self.numero_de_venta = numero_de_venta
        self.precio = precio
        self.tipo_pago = tipo_pago

def cargar(ventas):
    numero_de_venta = 0
    comodin = int(input("Elija:\nCargar = 1\nSALIR = 0\n"))
    if comodin == 1:
        bol = True
    elif comodin == 0:
        bol = False
    while bol != False:
        numero_de_venta =+ 1
        cod_producto = int(input("Ingresecodigo de producto:\nBANANA $100 = 1\nYOGURT $150 = 2\nCARNE $200 = 3\n"))
        if cod_producto == 1:
            tipo_producto = "BANANA"
            precio = 100
        elif cod_producto == 2:
            tipo_producto = "YOGURT"
            precio = 150
        elif cod_producto == 3:
            tipo_producto = "CARNE"
            precio = 200
        aux1 = int(input("Ingrese tipo de venta:\nEfectivo = 1\nTarjeta = 2\n"))
        if aux1 == 1:
            tipo_pago = "Efectivo"
        elif aux1 == 2:
            tipo_pago = "Tarjeta"
        ventas.append(venta(tipo_producto,numero_de_venta,precio,tipo_pago))
        comodin = int(input("Elija:\nCargar = 1\nSALIR = 0\n"))
        if comodin == 1:
            bol = True
        elif comodin == 0:
            bol = False
    
def mostrar(ventas):
    for i in range(len(ventas)):
        print("Numero venta: {}\nCodigo producto: {}\nPrecio: {}\nTipo de pago: {}" .format(
            ventas[i].numero_de_venta, ventas[i].tipo_producto,
            ventas[i].precio , ventas[i].tipo_pago))

def mostrar_mas_vendido(ventas):
    b = 0
    y = 0
    c = 0
    for i in range(len(ventas)):
        if ventas[i].tipo_producto == "BANANA":
            b += 1
        elif ventas[i].tipo_producto == "YOGURT":
            y += 1
        elif ventas[i].tipo_producto == "CARNE":
            c += 1
    if c > b and c > y:
        print("El producto mas vendido fue la CARNE")
    elif b > c and b > y:
        print("El producto mas vendido fue la BANANA")
    else:
        print("El producto mas vendido el YOGURT")

def mostrar_promedio_venta(ventas):
    acumulador = 0
    for i in range(len(ventas)):
        acumulador += ventas[i].precio
    promedio = acumulador/len(ventas)
    print("El promedio por venta fue de {} pesos por venta" .format(promedio))
    
def mostrar_venta_en_efectivo(ventas):
    acumulador = 0
    for i in range(len(ventas)):
        if ventas[i].tipo_pago == "Efectivo":
            acumulador += 1
    print("La cantidad de ventas en efectivo fue de {}".format(acumulador))
main()
```
